```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
```

```python
#S&P 500(^GSPC) // Type: Stock Market
tickerStrings = ["^NYA"]
```

```python
#Building our Dataframe
#I use the following function to create our Dataframe with the tickets I want
df_list = list()
for ticker in tickerStrings:
    data = yf.download(ticker, start="2022-01-01", end="2022-12-28", group_by="Column")
    data['ticker'] = ticker  # add this column because the dataframe doesn't contain a column with the ticker
    df_list.append(data)

# combine all dataframes into a single dataframe
df0 = pd.concat(df_list)
```

```python
df0 = df0[['Close']]
```

```python
df0.rename(columns={'Close': 'NYSE'}, inplace=True)
df0
```
